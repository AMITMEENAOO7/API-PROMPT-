# Using NLP

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.feature_extraction.text import TfidfVectorizer

from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

from sklearn.naive_bayes import MultinomialNB
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.pipeline import make_pipeline
from sklearn.ensemble import RandomForestClassifier
```

```python
# Sample data
emails = df.dropna().Description

# Labels corresponding to the categories (0, 1, 2 for example)
labels = df.dropna().Component

# Convert text data to feature vectors using BoW
vectorizer_cv = CountVectorizer()
Xcv = vectorizer_cv.fit_transform(emails)

# Split the data into training and testing sets
Xcv_train, Xcv_test, ycv_train, ycv_test = train_test_split(Xcv, labels, test_size=0.2, random_state=42)

# Train a Naive Bayes classifier
nb_cv = MultinomialNB()
nb_cv.fit(Xcv_train, ycv_train)

# Make predictions on the test set
nb_pred_cv = nb_cv.predict(Xcv_test)

# Evaluate the accuracy
nb_accuracy_cv = accuracy_score(ycv_test, nb_pred_cv)
print("Accuracy:", nb_accuracy_cv)
```

```python
for x, y in zip(ycv_test, nb_pred_cv):
    print(x, y)
```

```python
# Convert text data to feature vectors using BoW
vectorizer_tf = TfidfVectorizer()
Xtf = vectorizer_tf.fit_transform(emails)

# Split the data into training and testing sets
Xtf_train, Xtf_test, ytf_train, ytf_test = train_test_split(Xtf, labels, test_size=0.2, random_state=42)

# Train a Naive Bayes classifier
nb_tf = MultinomialNB()
nb_tf.fit(Xtf_train, ytf_train)

# Make predictions on the test set
nb_pred_tf = nb_tf.predict(Xtf_test)

# Evaluate the accuracy
nb_accuracy_tf = accuracy_score(ytf_test, nb_pred_tf)
print("Accuracy:", nb_accuracy_tf)
```

```python
for x, y in zip(ytf_test, nb_pred_tf):
    print(x, y)
```

```python
# Train a Logistic Regression classifier

from sklearn.linear_model import LogisticRegression

lr_cv = LogisticRegression()
lr_cv.fit(Xcv_train, ycv_train)

lr_pred_cv = lr_cv.predict(Xcv_test)

lr_accuracy_cv = accuracy_score(ycv_test, lr_pred_cv)
print("Accuracy:", lr_accuracy_cv)

for x, y in zip(ycv_test, lr_pred_cv):
    print(x, y)
```

```python
def Classifier_Results(classifier, X_train, X_test, y_train, y_test, vectorizer):
    classifier.fit(X_train, y_train)
    
    y_pred = classifier.predict(X_test)

    print(f"Accuracy for {vectorizer} vectorized {classifier} is {accuracy_score(y_test, y_pred)}\n")
    
    for x, y in zip(y_test, y_pred):
        print(x, y)
    print('\n\n')
```

```python
Classifier_Results(MultinomialNB(), Xcv_train, Xcv_test, ycv_train, ycv_test, 'CV')
Classifier_Results(MultinomialNB(), Xtf_train, Xtf_test, ytf_train, ytf_test, 'TF')
```

```python
Classifier_Results(LogisticRegression(), Xcv_train, Xcv_test, ycv_train, ycv_test, 'CV')
Classifier_Results(LogisticRegression(), Xtf_train, Xtf_test, ytf_train, ytf_test, 'TF')
```

```python
Classifier_Results(SVC(), Xcv_train, Xcv_test, ycv_train, ycv_test, 'CV')
Classifier_Results(SVC(), Xtf_train, Xtf_test, ytf_train, ytf_test, 'TF')
```

```python
Classifier_Results(RandomForestClassifier(n_estimators=100, random_state=42), Xcv_train, Xcv_test, ycv_train, ycv_test, 'CV')
Classifier_Results(RandomForestClassifier(n_estimators=100, random_state=42), Xtf_train, Xtf_test, ytf_train, ytf_test, 'TF')
```

# Clustering

```python
import numpy as np
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.cluster import KMeans
from sklearn.metrics import adjusted_rand_score, 
from sklearn.metrics.cluster import contingency_matrix
```

```python
def purity_score(y_true, y_pred):
    # compute contingency matrix (also called confusion matrix)
    contingency = contingency_matrix(y_true, y_pred)
    # return purity
    return np.sum(np.amax(contingency, axis=0)) / np.sum(contingency) 
```

```python
def km(X):
    # Apply K-Means clustering
    k = 3  # Number of clusters
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(X)

    # Assign cluster labels to emails
    cluster_labels = kmeans.labels_

    # Print the cluster assignments for each email
    for i, comp in enumerate(df.Component):
        print(f"Email {i + 1} with Component {comp} is assigned to Cluster {cluster_labels[i]}")

    # Evaluate the clustering performance (if ground truth labels are available)
    ground_truth_labels = df.Component  # Example ground truth labels
    
    print(f"Adjusted Rand Index: {adjusted_rand_score(ground_truth_labels, cluster_labels)}")
    print(f'Purity Score: {purity_score(ground_truth_labels, cluster_labels)}')
```
